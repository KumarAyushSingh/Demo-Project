# **Exploratory Data Analysis**

> Now let's begin some exploratory data analysis.

> We will start by checking out missing data first.







# **Missing Data**

> We can use seaborn to create a simple heatmap to see where we are missing data.



```
cars.isnull()
```

```
sns.heatmap(cars.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

**Observation 1**

> As we can see from the above heatmap that most of our null values are present in the 'New_Price' section and some of the null values are in 'Location','Engine', and ect.





Now lets visualize some more of the data.

```
sns.set_style('whitegrid')
sns.countplot(x='Fuel_Type',data=cars)
```

As we can observe that theres a huge demand of Diesel and Petrol cars as compared to CNG and LPG cars

```
sns.set_style('whitegrid')
sns.countplot(x='Owner_Type',data=cars)
```

And we can also observe that First hand cars are being more demanded as compared to Second and Third hand cars.

```
sns.set_style('whitegrid')
sns.countplot(x='Fuel_Type',hue='Owner_Type',data=cars,palette='RdBu_r')
```

In the above graph we have combined both of the graphs for better understanding

```
cars['Price'].hist(color='green',bins=40,figsize=(8,4))
```

**Observation 2**

> As we can see from the above graph that first hand cars are being sold for the maximum price and as the car gets older and older its resail value decreases



# **Now we will check for Corelations**

```
cars.corr()
```

```
plt.figure(figsize=(10,8))
sns.heatmap(cars.corr(), annot=True,mask=np.triu(cars.corr(),+1));
```

# **Observation 3**
